```
import os
data_dir = '/home/devang/Projects/PilotCrew/Data-Science-Bench/tasks/task_20/data'
os.listdir(data_dir)
```

```
import pandas as pd
harvests = pd.read_csv(os.path.join(data_dir, 'harvests.csv'))
plantings = pd.read_csv(os.path.join(data_dir, 'plantings.csv'))
fields = pd.read_csv(os.path.join(data_dir, 'fields.csv'))
print("Harvests:", harvests.columns.tolist())
print("Plantings:", plantings.columns.tolist())
print("Fields:", fields.columns.tolist())
```

```
df = fields.merge(plantings, on='field_id', how='left').merge(harvests, on='field_id', how='left')
print(len(fields), len(df))
df.head()
```

```
df['harvest_date'] = pd.to_datetime(df['harvest_date'])
df['target_harvest_date'] = pd.to_datetime(df['target_harvest_date'])
df['harvested_late'] = df['harvest_date'] > df['target_harvest_date']

# Calculate median yield per crop
median_yields = df.groupby('crop')['yield_tons'].transform('median')
df['below_median_yield'] = df['yield_tons'] < median_yields

# Both conditions
df['late_and_below_median'] = df['harvested_late'] & df['below_median_yield']

share = df['late_and_below_median'].mean()
print(round(share, 1))
```
