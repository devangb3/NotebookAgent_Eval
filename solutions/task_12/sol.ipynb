# task_12 Solution

```python

from pathlib import Path
import pandas as pd
import numpy as np

base = Path("../data")
```

```python

employees = pd.read_csv(base / "employees.csv")
projects = pd.read_csv(base / "projects.csv")
reviews = pd.read_csv(base / "reviews.csv")
timesheets = pd.read_csv(base / "timesheets.csv")
trainings = pd.read_csv(base / "trainings.csv")

ts = timesheets.merge(projects[["project_id", "client_billable"]], on="project_id")
billable = ts[ts["client_billable"] == 1].groupby("employee_id")["hours"].sum()
total = ts.groupby("employee_id")["hours"].sum()
q1 = str((billable / total).sort_values(ascending=False).index[0])

trained = set(trainings.groupby("employee_id")["training_code"].nunique().loc[lambda s: s > 2].index)
proj4 = set(timesheets.groupby("employee_id")["project_id"].nunique().loc[lambda s: s >= 4].index)
eligible = trained & proj4
q2 = f"{reviews[reviews['employee_id'].isin(eligible)]['review_score'].mean():.1f}"

employee_week = timesheets.groupby(["week_start", "employee_id"])["hours"].sum().reset_index(name="weekly_hours")
employee_week = employee_week.merge(employees[["employee_id", "department"]], on="employee_id")
q3 = str(employee_week.groupby("department")["weekly_hours"].mean().sort_values(ascending=False).index[0])

t1 = set(trainings[trainings["training_code"] == "T1"]["employee_id"])
t3 = set(trainings[trainings["training_code"] == "T3"]["employee_id"])
low_review = set(reviews[reviews["review_score"] < 3.5]["employee_id"])
q4 = str(len((t1 & t3) & low_review))

project_load = timesheets.groupby("project_id").agg(total_hours=("hours", "sum"), employees=("employee_id", "nunique"))
project_load["avg_per_employee"] = project_load["total_hours"] / project_load["employees"]
q5 = str(project_load["avg_per_employee"].sort_values(ascending=False).index[0])
```

Q1: Which employee_id has the highest billable-hours share, defined as hours on client_billable projects divided by total hours?

```python
print(q1)
```

Q2: What is the average review_score for employees who completed more than 2 distinct trainings and worked on at least 4 distinct projects, rounded to 1 decimal place?

```python
print(q2)
```

Q3: Which department has the highest average weekly total hours per employee-week?

```python
print(q3)
```

Q4: How many employees completed both T1 and T3 and also have review_score < 3.5?

```python
print(q4)
```

Q5: Which project_id has the highest average total hours per assigned employee over the full period?

```python
print(q5)
```
