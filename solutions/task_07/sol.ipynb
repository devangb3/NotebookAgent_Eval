# task_07 Solution

```python

from pathlib import Path
import pandas as pd

base = Path("../../tasks/task_07/data/")
```

```python

flights = pd.read_csv(base / "flights.csv")
weather = pd.read_csv(base / "airport_weather.csv")

flights["flight_date"] = pd.to_datetime(flights["flight_date"])
weather["flight_date"] = pd.to_datetime(weather["flight_date"])
```

Q1: Considering only non-cancelled flights, which origin-destination route has the highest average arrival delay?

```python
non_cancelled = flights[flights["cancelled"] == 0].copy()
non_cancelled["route"] = non_cancelled["origin"] + "-" + non_cancelled["dest"]
q1 = str(non_cancelled.groupby("route")["arr_delay_min"].mean().sort_values(ascending=False).index[0])
print(q1)
```

Q5: On non-cancelled flights whose origin airport had more than 5 mm rain, which origin airport_code has the highest average departure delay?

```python
rain_merge = non_cancelled.merge(weather[["flight_date", "airport_code", "rain_mm"]], left_on=["flight_date", "origin"], right_on=["flight_date", "airport_code"])
q5 = str(rain_merge[rain_merge["rain_mm"] >= 5].groupby("origin")["dep_delay_min"].mean().sort_values(ascending=False).index[0])
print(q5)
```
