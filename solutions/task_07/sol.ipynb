# task_07 Solution

```python

from pathlib import Path
import pandas as pd
import numpy as np

base = Path("../data")
```

```python

flights = pd.read_csv(base / "flights.csv")
weather = pd.read_csv(base / "airport_weather.csv")

flights["flight_date"] = pd.to_datetime(flights["flight_date"])
weather["flight_date"] = pd.to_datetime(weather["flight_date"])

non_cancelled = flights[flights["cancelled"] == 0].copy()
non_cancelled["route"] = non_cancelled["origin"] + "-" + non_cancelled["dest"]
q1 = str(non_cancelled.groupby("route")["arr_delay_min"].mean().sort_values(ascending=False).index[0])

wind_merge = flights.merge(weather[["flight_date", "airport_code", "wind_kph"]], left_on=["flight_date", "origin"], right_on=["flight_date", "airport_code"])
q2 = str(wind_merge[(wind_merge["wind_kph"] >= 25) & (wind_merge["cancelled"] == 1)].groupby("carrier").size().sort_values(ascending=False).index[0])

delayed = non_cancelled[non_cancelled["dep_delay_min"] > 5]
q3 = f"{(100 * (delayed['arr_delay_min'] <= 0).mean()):.1f}%"

total_distance = non_cancelled.groupby("flight_date")["distance_miles"].sum()
q4 = total_distance[total_distance == total_distance.max()].index.min().date().isoformat()

rain_merge = non_cancelled.merge(weather[["flight_date", "airport_code", "rain_mm"]], left_on=["flight_date", "origin"], right_on=["flight_date", "airport_code"])
q5 = str(rain_merge[rain_merge["rain_mm"] >= 5].groupby("origin")["dep_delay_min"].mean().sort_values(ascending=False).index[0])
```

Q1: Considering only non-cancelled flights, which origin-destination route has the highest average arrival delay?

```python
print(q1)
```

Q2: On days where the origin airport had wind_kph >= 25, which carrier had the most cancelled flights?

```python
print(q2)
```

Q3: Among non-cancelled flights with dep_delay_min > 5, what percentage arrived on time or early (arr_delay_min <= 0), rounded to 1 decimal place?

```python
print(q3)
```

Q4: Using only non-cancelled flights, which is the earliest flight_date among the date or dates with the highest total distance_miles flown?

```python
print(q4)
```

Q5: On non-cancelled flights whose origin airport had rain_mm >= 5, which origin airport_code has the highest average departure delay?

```python
print(q5)
```
