# task_18 Solution

```python

from pathlib import Path
import pandas as pd

base = Path("../../tasks/task_18/data/")
```

```python

books = pd.read_csv(base / "books.csv")
loans = pd.read_csv(base / "loans.csv")

loans["due_date"] = pd.to_datetime(loans["due_date"])
loans["return_date"] = pd.to_datetime(loans["return_date"])
loans["days_overdue"] = (loans["return_date"] - loans["due_date"]).dt.days.clip(lower=0)
loans["fine_amount"] = 0.5 * loans["days_overdue"]

```

Q2: Among members with at least 3 loans, which member_id has the highest average overdue days?

```python
merged = loans.merge(books[["book_id", "genre", "branch"]], on="book_id")
loan_counts = merged.groupby("member_id")["loan_id"].nunique()
eligible = set(loan_counts[loan_counts >= 3].index)
q2 = str(merged[merged["member_id"].isin(eligible)].groupby("member_id")["days_overdue"].mean().sort_values(ascending=False).index[0])
print(q2)
```
