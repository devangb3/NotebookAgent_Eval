# task_18 Solution

```python

from pathlib import Path
import pandas as pd
import numpy as np

base = Path("../data")
```

```python

books = pd.read_csv(base / "books.csv")
loans = pd.read_csv(base / "loans.csv")

loans["due_date"] = pd.to_datetime(loans["due_date"])
loans["return_date"] = pd.to_datetime(loans["return_date"])
loans["days_overdue"] = (loans["return_date"] - loans["due_date"]).dt.days.clip(lower=0)
loans["fine_amount"] = 0.5 * loans["days_overdue"]

merged = loans.merge(books[["book_id", "genre", "branch"]], on="book_id")
q1 = str(merged.groupby("genre")["days_overdue"].mean().sort_values(ascending=False).index[0])

loan_counts = merged.groupby("member_id")["loan_id"].nunique()
eligible = set(loan_counts[loan_counts >= 3].index)
q2 = str(merged[merged["member_id"].isin(eligible)].groupby("member_id")["days_overdue"].mean().sort_values(ascending=False).index[0])

q3 = f"{(100 * (loans['return_date'].dt.dayofweek >= 5).mean()):.1f}%"

merged["return_month"] = merged["return_date"].dt.to_period("M").astype(str)
q4 = str(merged.groupby("return_month")["fine_amount"].sum().sort_values(ascending=False).index[0])

loan_totals = merged.groupby("branch")["loan_id"].nunique()
book_totals = books.groupby("branch")["book_id"].nunique()
q5 = str((loan_totals / book_totals).sort_values(ascending=False).index[0])
```

Q1: Which genre has the highest average days_overdue?

```python
print(q1)
```

Q2: Among members with at least 3 loans, which member_id has the highest average days_overdue?

```python
print(q2)
```

Q3: What percentage of loans were returned on a Saturday or Sunday, rounded to 1 decimal place?

```python
print(q3)
```

Q4: Which return_date month has the highest total fine_amount?

```python
print(q4)
```

Q5: Which branch has the highest circulation turnover, defined as total loans of books from that branch divided by number of books assigned to that branch?

```python
print(q5)
```
