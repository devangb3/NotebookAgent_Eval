# task_02 Solution

```python

from pathlib import Path
import pandas as pd

base = Path("../../tasks/task_02/data/")
```

```python
stations = pd.read_csv(base / "stations.csv")
trips = pd.read_csv(base / "trips.csv")
weather = pd.read_csv(base / "weather.csv")
```

Q1: How many total ride minutes were taken by casual riders on days with rain_mm >= 6?

```python
trips["trip_date"] = pd.to_datetime(trips["trip_date"])
weather["trip_date"] = pd.to_datetime(weather["trip_date"])

merged = trips.merge(weather, on="trip_date")
q1 = str(int(merged.loc[(merged["rider_type"] == "casual") & (merged["rain_mm"] >= 6), "duration_min"].sum()))
print(q1)
```

Q2: During the 07:00 to 09:59 start-hour window, which station_id has the highest net inflow defined as arrivals minus departures?

```python
window = trips[(trips["start_hour"] >= 7) & (trips["start_hour"] <= 9)]
arrivals = window.groupby("end_station").size()
departures = window.groupby("start_station").size()
net_inflow = arrivals.sub(departures, fill_value=0).sort_values(ascending=False)
q2 = str(net_inflow.index[0])
print(q2)
```

Q3: What is the median trip duration, in minutes, for trips whose start and end stations are in different neighborhoods?

```python
station_nb = stations[["station_id", "neighborhood"]]
cross_nb = (
    trips.merge(station_nb.rename(columns={"station_id": "start_station", "neighborhood": "start_nb"}), on="start_station")
    .merge(station_nb.rename(columns={"station_id": "end_station", "neighborhood": "end_nb"}), on="end_station")
)
q3 = str(int(cross_nb.loc[cross_nb["start_nb"] != cross_nb["end_nb"], "duration_min"].median()))
print(q3)
```

Q4: Which trip_date has the highest average trip duration?

```python
avg_duration = trips.groupby("trip_date", as_index=False)["duration_min"].mean().sort_values(["duration_min", "trip_date"], ascending=[False, True])
q4 = avg_duration.iloc[0]["trip_date"].date().isoformat()
print(q4)
```

Q5: On days with temp_c >= 20, which destination neighborhood receives the most trip arrivals?

```python

hot_days = trips.merge(weather[["trip_date", "temp_c"]], on="trip_date")
hot_days = hot_days[hot_days["temp_c"] >= 20]
hot_days = hot_days.merge(station_nb.rename(columns={"station_id": "end_station", "neighborhood": "end_nb"}), on="end_station")
q5 = str(hot_days.groupby("end_nb").size().sort_values(ascending=False).index[0])
print(q5)
```
