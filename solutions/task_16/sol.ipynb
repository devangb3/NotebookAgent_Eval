# task_16 Solution

```python

from pathlib import Path
import pandas as pd
import numpy as np

base = Path("../data")
```

```python

bills = pd.read_csv(base / "bills.csv")
customers = pd.read_csv(base / "customers.csv")
usage = pd.read_csv(base / "usage.csv")

bills["due_date"] = pd.to_datetime(bills["due_date"])
bills["paid_date"] = pd.to_datetime(bills["paid_date"])

q1 = str(bills.groupby("plan_id")["overage_charge"].mean().sort_values(ascending=False).index[0])

usage_totals = usage.groupby("customer_id")[["weekend_data_gb", "weekday_data_gb"]].sum()
usage_totals["ratio"] = usage_totals["weekend_data_gb"] / usage_totals["weekday_data_gb"]
q2 = str(usage_totals["ratio"].sort_values(ascending=False).index[0])

exceeded = bills.assign(exceeded_voice=bills["voice_minutes"] > bills["included_minutes"]).groupby("customer_id")["exceeded_voice"].sum()
late = bills.groupby("customer_id").apply(lambda g: (g["paid_date"] > g["due_date"]).any(), include_groups=False)
qualified = (exceeded >= 2) & late
q3 = f"{(100 * qualified.mean()):.1f}%"

region_bills = bills.merge(customers[["customer_id", "region"]], on="customer_id")
q4 = str(region_bills.groupby("region")["bill_amount"].mean().sort_values(ascending=True).index[0])

q5 = str(bills.groupby("month")["overage_charge"].sum().sort_values(ascending=False).index[0])
```

Q1: Which plan_id has the highest average overage_charge per customer-month?

```python
print(q1)
```

Q2: Which customer_id has the highest ratio of total weekend_data_gb to total weekday_data_gb over the full period?

```python
print(q2)
```

Q3: What percentage of customers both exceeded included voice minutes in at least 2 months and paid late at least once, rounded to 1 decimal place?

```python
print(q3)
```

Q4: Which region has the lowest average bill_amount?

```python
print(q4)
```

Q5: Which month has the highest total overage_charge?

```python
print(q5)
```
